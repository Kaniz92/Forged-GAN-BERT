```python
def get_bleu_scores(generated_text, reference_text):
  smoothie = SmoothingFunction().method4

  ref_tokenized = word_tokenize(reference_text)
  gen_tokenized = word_tokenize(generated_text)

  bleu_score = sentence_bleu([ref_tokenized], gen_tokenized, smoothing_function=smoothie)

  return bleu_score
```

```python
def get_rouge_scores(generated_text, reference_text):
    rouge = Rouge()

    rouge_scores = rouge.get_scores(generated_text, reference_text[:len(generated_text)], avg=True)
    rouge1_score = rouge_scores["rouge-1"]["f"]
    rouge2_score = rouge_scores["rouge-2"]["f"]
    rougeL_score = rouge_scores["rouge-l"]["f"]

    return rouge1_score, rouge2_score, rougeL_score
```

```python
def get_meteor_scores(generated_text, reference_text):
    reference_text = word_tokenize(reference_text)
    generated_text = word_tokenize(generated_text)

    meteor_score_val = meteor_score([reference_text], generated_text)

    return meteor_score_val
```

```python
def jaccard_similarity(generated_text, reference_text):
    generated_text_tokens = set(word_tokenize(generated_text.lower()))
    reference_text_tokens = set(word_tokenize(reference_text.lower()))

    intersection = len(reference_text_tokens.intersection(generated_text_tokens))
    union = len(reference_text_tokens.union(generated_text_tokens))
    similarity = intersection / union

    return similarity
```

```python
def overlap_coefficient(generated_text, reference_text):
    reference_text_tokens = set(nltk.word_tokenize(reference_text.lower()))
    generated_text_tokens = set(nltk.word_tokenize(generated_text.lower()))

    intersection = len(reference_text_tokens.intersection(generated_text_tokens))
    min_length = min(len(reference_text_tokens), len(generated_text_tokens))
    similarity = intersection / min_length

    return similarity
```

```python
def calculate_cosine_similarity(generated_text, reference_text):
    vectorizer = CountVectorizer().fit_transform([generated_text, reference_text])
    vectors = vectorizer.toarray()
    return cosine_similarity(vectors)[0][1]
```

```python
def get_similarity_scores(row):
  generated_text = row['ChatGPTText']
  reference_text = row['BookText']

  row['BLEU'] = get_bleu_scores(generated_text, reference_text)
  row['ROUGE1'], row['ROUGE2'], row['ROUGEL'] = get_rouge_scores(generated_text, reference_text)
  row['METEOR'] = get_meteor_scores(generated_text, reference_text)
  row['Jaccard'] = jaccard_similarity(generated_text, reference_text)
  row['OverlapCoeff'] = overlap_coefficient(generated_text, reference_text)
  row['Cosine'] = calculate_cosine_similarity(generated_text, reference_text)

  return row
```

```python
def calculate_df(dataset_file):
    dataset = load_dataset('Authorship/ChatGPT', data_files=dataset_file)
    data = pd.DataFrame(dataset['train'])

    data = data.apply(lambda row: get_similarity_scores(row), axis=1)
    data.to_csv(f'/content/ChatGPT/similarity/{dataset_file}', index=False)
```
